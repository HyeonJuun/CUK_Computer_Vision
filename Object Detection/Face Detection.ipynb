```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
imagePath = './data/faces1.jpg'
cascPath = './data/haarcascade_frontalface_default.xml'

# Create the haar cascade
faceCascade = cv2.CascadeClassifier(cascPath)

# Read the image
image = cv2.imread(imagePath)
imageRGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Detect facse in the image
faces = faceCascade.detectMultiScale(gray,
                                   minNeighbors=5,  #    ()
                                   minSize=(20, 20) #   
                                    ) 


#     
for (x, y, w, h) in faces:
    cv2.rectangle(imageRGB, (x, y), (x+w, y+h), (255, 0, 0), 2)
    
#     .
plt.figure(figsize=(15, 15))
plt.imshow(imageRGB)
```

```python
def FaceDetection(image):
    faceimage = image.copy()
    
    # Create the haar cascade
    faceCascade = cv2.CascadeClassifier(cascPath)
    
    # Read the image
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Detect faces in the image
    faces = faceCascade.detectMultiScale(gray,minNeighbors=5,minSize=(20, 20)) 
    
    # draw lane
    for (x, y, w, h) in faces:
        cv2.rectangle(imageRGB, (x, y), (x+w, y+h), (255, 0, 0), 2)
    
    return faceimage
```

```python
cap = cv2.VideoCapture('./data/siliconvalley.mp4')
frame_size = (int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)), int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)))
print('frame_size = ',frame_size)

while True:
    retval, frame = cap.read() #  
    if not retval:
        break
        
    # lane detection
    face = FaceDetection(frame)
    cv2.imshow('face', face)
    
    key = cv2.waitKey(25)
    if key == 27: # Esc
        break
        
if cap.isOpened():
    cap.release()
    
cv2.destroyAllWindows()
```
